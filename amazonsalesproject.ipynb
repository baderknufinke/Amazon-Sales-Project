```python
# drop any rows that have null values. Because we are not sure of the results, let's not mess with the current df and create a new one to mess with.

sales_data_dropped = sales_data.dropna()
```

```python
# the previous dropped too many amounts. now we will only drop those with missing data from one column - the 'amounts' column. 

sales_data_cleaned = sales_data.dropna(subset = ['Amount'])
```

```python
print(sales_data_cleaned.isnull().sum())
```

SLICING AND FILTERING DATA

```python
""" 1. How manysales have tehy made with amounts more than 1000 """

high_amount_sales = sales_data_cleaned[sales_data_cleaned['Amount'] > 1000]
```

```python

# selecting a subset from the column category 
category_data = sales_data_cleaned[sales_data_cleaned['Category'] == 'Top']
```

```python
""" 2. how many sales have they made that belong to the Category 'Tags' and have a Quantity of 3 """
# Selecting a subset from the data based on multiple conditions
filtered_data = sales_data_cleaned[(sales_data_cleaned['Category'] == 'Top') & (sales_data_cleaned['Qty'] == 3)]
```

AGGREGATING DATA

```python
""""" 3. The total sales by Category """""

category_totals = sales_data_cleaned.groupby('Category', as_index=False)['Amount'].sum()
category_totals = category_totals.sort_values('Amount', ascending=False)
```

```python
"""4. Average Amount by Category and Fulfilment"""
fulfilment_averages = sales_data_cleaned.groupby(['Category', 'Fulfilment'], as_index=False)['Amount'].mean().round(2)
fulfilment_averages = fulfilment_averages.sort_values('Amount', ascending=False)
```

```python
"""4a. Average Amount by Category and status"""
status_averages = sales_data_cleaned.groupby(['Category', 'Status'], as_index=False)['Amount'].mean().round(2)
status_averages = status_averages.sort_values('Amount', ascending=False)
```

```python
"""5. Total Sales by Fullfillment and Shipment Type"""
shipment_and_Fulfilment_totals = sales_data_cleaned.groupby(['Shipment', 'Fulfilment'], as_index=False)['Amount'].sum().round(2)
shipment_and_Fulfilment_totals  = shipment_and_Fulfilment_totals .sort_values('Amount', ascending=False)
# the shipment column does not exist.
```

```python
#  Let's see what other column we can use. Let's use courier status

sales_data_cleaned.columns
```

```python
"""5a. Total Sales by Fullfillment and Shipment"""
shipment_and_Fulfilment_totals = sales_data_cleaned.groupby(['Fulfilment', 'Courier Status'], as_index=False)['Amount'].sum().round(2)
shipment_and_Fulfilment_totals = shipment_and_Fulfilment_totals.sort_values('Amount', ascending=False)
""" change the column name courier status to shipment"""
shipment_and_Fulfilment_totals.rename(columns ={'Courier Status': 'Shipment'}, inplace=True) # we do not need to reassign the variable by using '='
```

EXPORTING THE DATA

```python
"""6. Export dataFrames created in 4 and 5a to excel"""

status_averages.to_excel('average_sales_by_cat_&_stat.xlsx', index=False)
shipment_and_Fulfilment_totals.to_excel('sales_by_Fullfillment_and_Shipment.xlsx', index=False)
```
